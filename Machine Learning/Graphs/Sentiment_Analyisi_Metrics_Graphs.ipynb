```python
import matplotlib.pyplot as plt
import numpy as np

# Performance metrics for each model
models = ['Linear Support Vector Machine', 'Random Forest', 'Logistic Regression','Naive Bayesian']
accuracy = [0.90,0.95, 0.89,0.84]
precision_class1 = [0.81, 0.92, 0.80,0.85]
precision_class2 = [0.81, 0.98, 0.62,0.66]
precision_class3 = [0.93, 0.95, 0.93,0.84]
recall_class1 = [0.87, 0.89, 0.86,0.57]
recall_class2 = [0.21, 0.61, 0.14,0.02]
recall_class3 = [0.97, 0.99, 0.97,0.99]
f1_score_class1 = [0.84, 0.91, 0.83,0.68]
f1_score_class2 = [0.33, 0.75, 0.23,0.03]
f1_score_class3 = [0.95, 0.97, 0.95,0.91]

# Plotting the metrics
fig1, axs1 = plt.subplots(1, 1, figsize=(12, 6))
index = np.arange(len(models))
bar_width = 0.3

fig2, axs2 = plt.subplots(1, 1, figsize=(12, 6))
index = np.arange(len(models))
bar_width = 0.3

fig3, axs3 = plt.subplots(1, 1, figsize=(10, 6))
index = np.arange(len(models))
bar_width = 0.3

# Accuracy subplot
axs1.bar(index, accuracy, bar_width)
axs1.set_xticks(index)
axs1.set_xticklabels(models)
axs1.set_ylabel('Accuracy')
axs1.set_title('Model Accuracy')

# Precision subplot
axs2.bar(index, precision_class1, bar_width, label='Negative')
axs2.bar(index + bar_width, precision_class2, bar_width, label='Neutral')
axs2.bar(index + 2 * bar_width, precision_class3, bar_width, label='Positive')
axs2.set_xticks(index + bar_width)
axs2.set_xticklabels(models)
axs2.set_ylabel('Precision')
axs2.set_title('Model Precision')
axs2.legend()

# Recall subplot
axs3.bar(index, recall_class1, bar_width, label='Negative')
axs3.bar(index + bar_width, recall_class2, bar_width, label='Neutral')
axs3.bar(index + 2 * bar_width, recall_class3, bar_width, label='Positive')
axs3.set_xticks(index + bar_width)
axs3.set_xticklabels(models)
axs3.set_ylabel('Recall')
axs3.set_title('Model Recall')
axs3.legend()

# Adjusting spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
