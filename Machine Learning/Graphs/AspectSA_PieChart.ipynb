```python
import matplotlib.pyplot as plt

# Data
labels = ['Positive', 'Neutral', 'Negative']
sizes = [50, 10, 40]
colors = ['#34ad65', '#f5c542', '#e74c3c']

# Create pie chart
plt.pie(sizes,colors=colors,labels=labels,autopct=lambda x: '{:.0f}'.format(x*sum(sizes)/100), startangle=90)
title_font = {'fontname': 'Arial', 'fontsize': 16, 'fontweight': 'bold'}
plt.title('Battery', **title_font)
# Add actual numbers


# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
```
